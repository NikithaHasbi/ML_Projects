```python
##missing values#
#you can get the # missing values in two ways
print(sum(np.isnan(titan.Survived)))
print(titan.info())
```

```python
## to delete the missing values in Embarked##
ind= pd.isna(titan.Embarked)
titan = titan[-ind]
```

```python
print(sum(np.isnan(titan.Survived)))
print(titan.info())
```

```python
# to impute the missing valeu
## to search for stuff in rows ##
# 5) to create a column and group them #
ind1 = titan['Name'].str.contains('Mr.')
ind2 = titan['Name'].str.contains('Mrs.')
ind3 = titan['Name'].str.contains('Miss.')
ind4 = titan['Name'].str.contains('Master.')
titan['title']=""
titan.title[ind1]=1
titan.title[ind2]=2
titan.title[ind3]=3
titan.title[ind4]=4
```

```python
### to impute the missing values in Age ## using Regex
imp = pd.isnull(titan.Age)
for index,row in titan.iterrows():
    if imp[index]==True:
        item = row["title"]
        titan.loc[index,"Age"]=(np.mean(titan.loc[titan.title==item,"Age"]))
```

```python

print(sum(np.isnan(titan.Survived)))
print(titan.info())
```

```python
titan.head(10)
```

```python
titan=titan.drop("Name", axis=1)
```

### Logisitc Regression ### 
#### Dummy Variable Creation ####

```python
cat_col = titan.select_dtypes(include=['object']).columns
dummies = pd.get_dummies(titan[cat_col],prefix= cat_col,drop_first=True)
titan.drop(cat_col,axis=1,inplace=True)
X= pd.concat([titan,dummies],axis=1)
```

```python
X.info()
```

```python
X.head(10)
```

```python
##divide the data into train and set## ### target variables are always answers, so we drop it because we have to only input the questions##
Y= X["Survived"]
X= X.drop("Survived",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, Y_train, Y_test =train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
##feeding data to the ML algorithm## ##start training the algorithm##
### always create an object of class 'algorithm' as this algorithm will contain all the necessary formulas ##
### and then you use fit() object to train the algorithm by giving it data so that it analyses all of it##
## lastly you give the algo the new test questions and store the answers in Y_pred ##
from sklearn.linear_model import LogisticRegression
log_reg= LogisticRegression()
log_reg.fit(X_train,Y_train)
Y_pred = log_reg.predict(X_test)
```

```python
## we have to now check if the Y_pred is equal to Y_ test and calc accuracy ##
## analysisng the error through metrics, must be 0##
from sklearn import metrics
metrics.log_loss(Y_test,Y_pred)
```

```python
## must be as close to 1 as possible, ranges from 0 to 1 ##
metrics.accuracy_score(Y_test,Y_pred)
```
